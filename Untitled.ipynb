```python
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'slope','exang', 'ca', 'thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
standardScaler = StandardScaler()
robustScaler = RobustScaler()
col_to_scale = ['age', 'trestbps',"chol", 'thalach', 'oldpeak']
dataset[col_to_scale] = robustScaler.fit_transform(dataset[col_to_scale])
```

```python
y = dataset['target']
x = dataset.drop(['target'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

params = {'n_neighbors' : list(range(1, 20)),
         'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'], 
         }

knn = KNeighborsClassifier(n_jobs=-1)
grid_knn = GridSearchCV(knn,param_grid=params,  scoring='accuracy', cv=10, n_jobs=-1, verbose=1)
```

```python
grid_knn.fit(X_train, y_train)
```

```python
grid_knn.best_score_
```

```python
grid_knn.best_params_
```

```python
grid_mean_scores = grid_knn.cv_results_["mean_test_score"]
plt.plot(list(range(1, 77)), grid_mean_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('mean test Accuracy')
```

```python
knn = KNeighborsClassifier(n_neighbors=8,algorithm='auto', n_jobs=-1)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1)

rf_randomSearch = RandomizedSearchCV(rf, param_distributions=random_grid, cv=5, n_jobs=-1,
                                     random_state=42,n_iter=10,verbose=2)
```

```python
rf_randomSearch.fit(X_train, y_train)
```

```python
rf_randomSearch.best_params_
```

```python
rf_randomSearch.best_score_
```

```python
rf = RandomForestClassifier(n_estimators=2000,min_samples_split=5,min_samples_leaf=2,
                            max_features='sqrt',max_depth=50,bootstrap=True)
rf.fit(x, y)
```

```python
rf.score(x, y)
```

```python
grid_mean_scores = rf_randomSearch.cv_results_["mean_test_score"]
plt.plot(list(range(1, 11)), grid_mean_scores)
plt.xlabel('iteration of Random Forests')
plt.ylabel('mean test Accuracy')
```
